```python
import os
import random
import shutil

data_path = "/kaggle/input/coco-2017-dataset/coco2017/val2017"

data_path2="/kaggle/working/coco/images"
# path to destination folders
train_folder = os.path.join(data_path2, 'train')
val_folder = os.path.join(data_path2, 'eval')
test_folder = os.path.join(data_path2, 'test')

# Define a list of image extensions
image_extensions = ['.jpg', '.jpeg', '.png', '.bmp']

# Create a list of image filenames in 'data_path'
imgs_list = [filename for filename in os.listdir(data_path) if os.path.splitext(filename)[-1] in image_extensions][0:6000]

# Sets the random seed 
random.seed(42)

# Shuffle the list of image filenames
random.shuffle(imgs_list)

# determine the number of images for each set
train_size = int(len(imgs_list) * 0.8)
# val_size = int(len(imgs_list) * 0.25)
test_size = int(len(imgs_list) * 0.2)

# Create destination folders if they don't exist
for folder_path in [train_folder, val_folder, test_folder]:
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

# Copy image files to destination folders
for i, f in enumerate(imgs_list):
    if i < train_size:
        dest_folder = train_folder
#     elif i < train_size + val_size:
#         dest_folder = val_folder
    else:
        dest_folder = test_folder
    shutil.copy(os.path.join(data_path, f), os.path.join(dest_folder, f))
```

```python
import pandas as pd
import os

# Load your dataset into a pandas DataFrame

# Create a directory to store the output files
output_directory = "/kaggle/working/coco/images/train"
os.makedirs(output_directory, exist_ok=True)

# Group the DataFrame by the 'image' column
grouped = filtered_test.groupby('image')

# Loop through each group
for name, group in grouped:
    # Create a text file for each image
    output_filename = os.path.join(output_directory, f"{name[:-4]}.txt")
    with open(output_filename, "w") as file:
        # Write the information for each row in the group
        for _, row in group.iterrows():
            info = f"{row['category_id']} {row['x1']} {row['y1']} {row['x2']} {row['y2']}"
            file.write(info + "\n")
```
